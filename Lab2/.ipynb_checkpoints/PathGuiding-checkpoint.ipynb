*最短路径算法实现

```python
from scipy.misc import imsave  
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python

start=(0,0)
target=(4,4)
map_list=[]
path=[]
visited=[]
f = open("data/0.1.txt", "r")
#将地图数据读进map中
for i in range(100):
    line=f.readline()
    row=list(line)#转化为数组
    row.pop()#去除尾部换行符
    map_list.append(row)
#print(map_list)
f.close()
```

```python
def draw_map(points):
    
    fig = plt.figure(figsize=(13,13)) 
    plt.scatter(x=[x[0] for x in points] ,y=[x[1] for x in points],c = 'r',marker = 's') 
    plt.gca().invert_yaxis() 
    plt.show() 
```

```python
barrier=[]
for index_x in range(100):
    for index_y in range(100):
        if map_list[index_x][index_y]=='1':
            barrier.append([index_y,index_x])
draw_map(barrier)
```
